```python
#Data preprocessing Step 1
#reading the csv file
import pandas as pd
dataset= pd.read_csv("Road_Accident.csv")
dataset
```

Step 2:
Dividing the dataset
x -------> input values
y--------> output values

```python
x=dataset.iloc[:,1:-1].values
x
```

```python
y=dataset.iloc[:,-1].values
y
```

Step 3 : Finding Empty cells, in the above dataset we didnot find any empty values
Step 4 : If any data is present in string format we should encode it
Step 5 : Splitting the data into training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

Step 6 : Applying Machine Learning algorithms to find which algorithm provides best accuracy

```python
#regression model Linear Regression
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#applying KNN
from sklearn.neighbors import KNeighborsClassifier
classifier1=KNeighborsClassifier()
classifier1.fit(x_train,y_train)
```

```python
#applying decision tree
from sklearn.tree import DecisionTreeClassifier
classifier2=DecisionTreeClassifier()
classifier2.fit(x_train,y_train)
```

```python
#apply Random Forest  ---- Bunch of trees
from sklearn.ensemble import RandomForestClassifier
classifier3=RandomForestClassifier(n_estimators=100)
classifier3.fit(x_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB  
classifier4 = GaussianNB()  
classifier4.fit(x_train, y_train)  
```

```python
from sklearn.metrics import r2_score
y_pred5=regressor.predict(x_test)
print(r2_score(y_pred5,y_test))
```

```python
#finding accuracy
from sklearn.metrics import accuracy_score
y_pred1=classifier1.predict(x_test)
y_pred2=classifier2.predict(x_test)
y_pred3=classifier3.predict(x_test)
y_pred4=classifier4.predict(x_test)

print(accuracy_score(y_pred1,y_test)*100)
print(accuracy_score(y_pred2,y_test)*100)
print(accuracy_score(y_pred3,y_test)*100)
print(accuracy_score(y_pred4,y_test)*100)
```
